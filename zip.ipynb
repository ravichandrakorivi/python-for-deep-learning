# `zip()` Function
- `zip()` is a built-in function in Python.
- It combines multiple iterables (lists, tuples, etc.) element-wise.
- Returns an iterator of tuples.

Syntax : `zip(iterable1, iterable2, ...)`

## Basic Example

```python
a  = [1, 2, 3]
b = ['x', 'y', 'z']

result = zip(a, b)
list(result)
```

```python
a  = (1, 2, 3)
b = ('x', 'y', 'z')
list(zip(a, b))
```

# Stops at the Shortest Iterable

```python
a = [1, 2, 3]
b = ['x', 'y']

list(zip(a,b))
```

## Works with Multiple Iterables

```python
a = [1, 2]
b = ['x', 'y']
c = [True, False]

list(zip(a, b, c))
```

## Using `zip()` in Loops

```python
names = ['A', 'B', 'C']
marks = [90, 80, 85]
for name, mark in zip(names, marks):
    print(name, mark)
```

## `zip()` returns an iterator, not a list
- `z = zip()` returns an iterator, not a list.
- Iterators can be used only once.
- `list(z)` consumes the iterator `z`.
- After this, `z` becomes empty.

This is a classic iterator exhaustion issue in Python.

```python
s = [3, 2, 1]
z = zip(s[:-1], s[1:])
```

```python
list(z)
```

```python
list(z)
```

## Converting zip to list

```python
z = zip([1, 2], ['a', 'b'])
z
```

```python
list(z)
```

## Converting zip to list

```python
z = zip([1, 2], ['a', 'b'])
```

```python
tuple(z)
```

## Converting zip to dictionary

```python
z = zip([1, 2], ['a', 'b'])
```

```python
dict(z)
```

## Unzipping

```python
z = zip([1, 2], ['a', 'b'])
```

```python
u = zip(*z)
```

```python
list(u)
```

## Common Use Cases

### Combining lists

```python
keys = ['name', 'age']
values = ['Ravi', 25]
d = dict(zip(keys, values))
d
```

### Parallel Iteration

```python
for i, j in zip([1,2,3], [4,5,6]):
    print(i + j, end=" ")
print("\n")
```
