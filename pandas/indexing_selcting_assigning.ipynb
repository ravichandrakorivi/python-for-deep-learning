# Pandas : Indexing, Selecting & Assigning

# 1. Index

- Every DataFrame has row labels (index) and column labels.
- Default index: 0, 1, 2, ....
- Index helps in label-based data access.

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "Name": ["A", "B", "C"],
    "Age": [25, 30, 22]
})
```

```python
df
```

```python
df.index
```

```python
df.index.to_list()
```

# 2. Selecting Columns from DataFrame

## Single column → Series

```python
df["Name"]
```

## Single column → Dataframe

```python
df[["Name"]]
```

## Multiple columns → DataFrame

```python
df[["Name", "Age"]]
```

# 3. Selecting Rows from DataFrame

## Using slicing

- This works like Python slicing.
- end excluded.

```python
df[0:2]
```

## `.loc[]` → Label-Based Selection

- Uses index labels.
- Includes end index.

```python
# first row
df.loc[0]
```

```python
# rows 0 to 2 (inclusive)
df.loc[0:2]
```

```python
# specific cell
df.loc[0, "Name"]
```

```python
# Multiple selections
df.loc[[0, 2], ["Name", "Age"]]
```

```python
# Multiple selections
df.loc[0:2, ["Name", "Age"]]
```

## `.iloc[]` → Position-Based Selection

- Uses integer positions.
- Works like Python indexing (end excluded).

```python
# first row
df.iloc[0]
```

```python
# rows 0 through to 2
df.iloc[0:3]
```

```python
# # specific cell : first row, second column
df.iloc[0, 1]
```

## `.loc` vs `.iloc`

| feature | `.loc` | `iloc` |
|:---|:---|:---|
| Based on | Labels | Positions |
| End index | Included | Excluded |

# 4. Boolean Indexing (Filtering)

```python
df["Age"] > 23
```

```python
df[df["Age"] > 23]
```

```python
(df["Age"] > 22) & (df["Name"] == "B")
```

```python
df[(df["Age"] > 22) & (df["Name"] == "B")]
```

# 5. Assigning / Modifying Data

## Assign New Column

```python
df["Salary"] = [50000, 60000, 45000]
```

```python
df
```

## Modify Existing Column

```python
df["Age"] = df["Age"] + 1
```

```python
df
```

## Conditional Assignment

```python
df.loc[df["Age"] > 25, "Senior"] = True
```

```python
df
```

```python
df.loc[df["Age"] > 30, "Salary"] = 70000
```

```python
df
```

## Assign Single Value

```python
df.loc[0, "Age"] = 27
```

```python
df
```

# 8. Examples

```python
df = pd.DataFrame({
    "City": ["Bangalore", "Chennai", "Bangalore", "Mumbai"],
    "Revenue": [100, 200, 150, 300]
})
df
```

```python
# Filter Bangalore data
df[df["City"] == "Bangalore"]
```

```python
# Increase revenue for Bangalore
df.loc[df["City"] == "Bangalore", "Revenue"] += 50
```
