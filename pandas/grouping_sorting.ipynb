# Pandas: Grouping & Sorting

# Grouping in Pandas (`groupby`)

Grouping means splitting data into categories based on column values. Then we can apply operations (sum, mean, count, etc.) on each group.

````python
import pandas as pd
````

````python
data = {
    "OrderID": [101,102,103,104,105,106,107,108,109,110,111,112,113,114,115],
    "Customer": ["Alice","Bob","Charlie","Alice","Bob","David","Eve","Frank","Alice","Eve","Charlie","Bob","David","Frank","Alice"],
    "City": ["Bangalore","Chennai","Mumbai","Bangalore","Chennai","Delhi","Mumbai","Delhi","Bangalore","Mumbai","Mumbai","Chennai","Delhi","Delhi","Bangalore"],
    "Category": ["Electronics","Clothing","Electronics","Furniture","Electronics","Clothing","Furniture","Electronics","Clothing","Electronics","Furniture","Clothing","Electronics","Furniture","Electronics"],
    "Product": ["Laptop","Shirt","Mobile","Table","Headphones","Jacket","Chair","Tablet","T-shirt","Mobile","Sofa","Jeans","Laptop","Chair","Mobile"],
    "Quantity": [1,2,1,1,2,1,4,1,3,2,1,2,1,2,1],
    "Price": [60000,1500,20000,7000,3000,2500,5000,15000,800,22000,12000,1800,65000,5500,21000],
    "PaymentMode": ["UPI","Card","Cash","UPI","Card","Cash","UPI","Card","UPI","Cash","Card","UPI","Cash","Card","UPI"]
}

df = pd.DataFrame(data)
df
````

````python
df["Revenue"] = df["Quantity"] * df["Price"]
````

````python
df
````

## Group by one column

````python
df.groupby("City")
````

## Apply aggregation

````python
# Average salary per department
df.groupby("City")["Revenue"].mean()
````

## Apply Multiple aggregations

````python
df.groupby("City")["Revenue"].agg(["min", "max", "mean", "sum", "count"])
````

````python
df.groupby("City")["Revenue"].agg(["mean"])
````

## Group by multiple columns

````python
df.groupby(["City", "PaymentMode"])["Revenue"].sum()
````

````python
df.groupby(["City", "PaymentMode"])["Revenue"].agg(["sum"])
````

## Reset index
- `reset_index()` converts the grouped result into a DataFrame.

````python
df.groupby("City")["Revenue"].mean().reset_index()
````

- Without `reset_index()`, we get a Series with City as index.

````python
df.groupby("City")["Revenue"].mean()
````

## Using `as_index=False`

````python
# as_index=False → returns a DataFrame
df.groupby("City", as_index=False)["Revenue"].mean()
````

````python
# default (as_index=True) → returns a Series with index
df.groupby("City", as_index=True)["Revenue"].mean()
# df.groupby("City")["Revenue"].mean()
````

## Apply custom function

````python
# Series
df.groupby("City")["Revenue"].apply(lambda x: x.max() - x.min())
````

For each City, it calculates the Revenue range (highest Revenue − lowest Revenue). 

```python
.apply(lambda x: x.max() - x.min())
```
- `x` = all Revenues in that department
- `x.max()` = highest Revenue
- `x.min()` = lowest Revenue
- `subtract` → gives range

````python
# Dataframe
df.groupby("City")["Revenue"].apply(lambda x: x.max() - x.min()).reset_index()
````

````python
# Dataframe
df.groupby("City", as_index=False)["Revenue"].apply(lambda x: x.max() - x.min())
````

````python
df.groupby("City")["Revenue"].agg(lambda x: x.max() - x.min())
````

````python
df.groupby("City")["Revenue"].agg(revenue_range=lambda x: x.max() - x.min())
````

# Sorting in Pandas (`sort_values`, `sort_index`)

````python
# Sorts in ascending order by default
df.sort_values("Revenue")
````

````python
# Sort descending
df.sort_values("Revenue", ascending=False)
````

````python
# Reset index after sorting
df.sort_values("Revenue", ascending=False).reset_index()
````

````python
# Reset index after sorting
df.sort_values("Revenue", ascending=False).reset_index(drop=True)
````

````python
# Sort by multiple columns
df.sort_values(["City", "Revenue"])
````

````python
# Different order for multiple columns
df.sort_values(["City", "Revenue"], ascending=[True, False])
````

## Grouping + Sorting Together

````python
df.groupby("City", as_index=False)["Revenue"].mean() \
.sort_values("Revenue", ascending=False)
````
