## `np.random.randn()` function

The `np.random.randn()` function in Python's NumPy library returns a sample or samples from the **standard normal distribution** (mean 0, variance 1).

- Distribution: Standard Normal or Standard Gaussian
    - Mean: 0
    - Std dev: 1
- Range: (-∞, ∞)
- Arguments: Shape as separate args

```python
# Return a single random float
np.random.randn()
```

```python
# Return a 1D array of random floats
np.random.randn(5)
```

```python
# Return a 2D array (matrix) of random floats
np.random.randn(3, 4)
```

```python
# Return a 3D array (tensor) of random floats
np.random.randn(3, 2, 1)
```

```python
sizes = [10, 5, 3]
```

```python
biases = [np.random.randn(size, 1) for size in sizes[1:]]
biases
```

```python
sizes = [3, 2, 1]
```

```python
weights = [np.random.rand(next, curr) for curr, next in zip(sizes[:-1], sizes[1:])]
weights
```

## `numpy.reshape()`

`numpy.reshape()` is used to change the shape (dimensions) of an array without changing its data.

- Changes shape (dimensions) of an array
- Does NOT change data
- Total elements remain constant
- Syntax: `numpy.reshape(array, new_shape)` or `array_name.reshape(new_shape)` 

```python
a = np.array([1, 2, 3, 4, 5, 6])
```

```python
a.reshape((2, 3))
```

```python
# NumPy can automatically calculate one dimension
a = np.array([1,2,3,4,5,6])
a.reshape((2, -1))

# -1 tells NumPy to figure out the length/size of other axis/dimension
```

## `numpy.random.shuffle()`

In NumPy, you can randomly shuffle an array using two primary methods: `numpy.random.shuffle()` to modify the array in-place or `numpy.random.permutation()` to return a new shuffled array while leaving the original intact.

The `numpy.random.shuffle()` function modifies the array directly and returns None. For multi-dimensional arrays, it only shuffles along the first axis (rows).

```python
# Use the recommended Generator instance for new code
rng = np.random.default_rng()
```

```python
arr_1d = np.array([1, 2, 3, 4, 5])
rng.shuffle(arr_1d) # Shuffles the array in-place
print("Shuffled 1D array:", arr_1d)
```

```python
arr_2d = np.array([[10, 20], [30, 40], [50, 60]])
rng.shuffle(arr_2d) # Only shuffles the rows (1st axis)
print("Shuffled 2D array (rows only):\n", arr_2d)
```

## `numpy.random.permutation()`

The `numpy.random.permutation()` function returns a new shuffled array (a copy) and keeps the original array unchanged. This is useful when you need to preserve the original data.

```python
# Use the recommended Generator instance for new code
rng = np.random.default_rng()
```

```python
original_arr = np.array([1, 2, 3, 4, 5])
shuffled_arr = rng.permutation(original_arr)
print("Original array:", original_arr)
print("New shuffled array:", shuffled_arr)
```

```python
arr_2d_org = np.array([[10, 20], [30, 40], [50, 60]])
arr_2d_shuf = rng.permutation(arr_2d_org) # Only shuffles the rows (1st axis)
print("Original array:", arr_2d_org)
print("New shuffled array:", arr_2d_shuf)
```

```python
# It can also take an integer N and shuffle np.arange(N)
shuffled_range = rng.permutation(5)
print("Shuffled range(5):", shuffled_range)
```

## `numpy.arange()`

The `numpy.arange()` function creates an array of evenly spaced values within a specified interval. It is similar to Python's built-in `range()` function but returns a NumPy array (ndarray) and can handle floating-point numbers.

The syntax for `np.arange()` is as follows, with parameters after `stop` being optional: 

`numpy.arange([start,] stop, [step,] dtype=None)`

**Parameters**:
- `start` (optional): The beginning of the interval. The default is 0. The interval includes this value.
- `stop`: The end of the interval. The interval excludes this value (half-open interval [`start`, `stop`)).
- `step` (optional): The spacing between consecutive values. The default step is 1. This can be a negative value to create a descending sequence.
- `dtype` (optional): The data type of the output array elements. If not specified, the data type is inferred from the other arguments. 

**Basic usage (only `stop` provided)**: 

Create an array starting at `0` up to (but not including) `5`, with a default step of `1`.

```python
np.arange(5)
```

**Specifying `start` and `stop`**: 

Create an array from `2` up to (but not including) `8`.

```python
np.arange(2, 8)
```

**Specifying `start`, `stop`, and `step`**: 

Create an array from `2` up to (but not including) `10`, with a step of `2`.

```python
np.arange(2, 10, 2)
```

**Using floating-point numbers**

```python
np.arange(1.0, 5.0, 0.5)
```

## `numpy.zeros()`

The `numpy.zeros()` function creates a new array of a specified shape and data type, with all elements initialized to zero. This function is commonly used for pre-allocating memory for arrays that will be populated with data later.

**Syntax**:

`numpy.zeros(shape, dtype=float64, order='C', like=None)`

- `shape` (required): An integer or a tuple of integers defining the dimensions of the array (e.g., `5` for a 1D array of 5 elements, or (`2`, `3`) for a 2D array with `2` rows and `3` columns).
- `dtype` (optional): The desired data type for the array elements. The default is `numpy.float64`. You can specify `int` to create an array of integers.
- `order` (optional): Specifies the memory layout, either C-style (row-major) or Fortran-style (column-major). The default is 'C'.

```python
# Create a 1D array
np.zeros(5)
```

```python
# Create a 2D array (matrix)
np.zeros((3, 4))
```

```python
# Create an array with integer data type
np.zeros((2, 3), dtype=int)
```

## `np.ones()`

`np.ones()` is a NumPy function used to create a new array of a specified shape and data type, filled entirely with ones. It is primarily used for initializing matrices or creating arrays for numerical computations. The default data type is `float64`, but it can be changed to integers or other types using the `dtype` parameter.

**Syntax**: 

`numpy.ones(shape, dtype=None, order='C')`

- `shape`: An integer or tuple of integers defining the dimensions (e.g., `(2, 3)` for a 2x3 matrix).
- `dtype` (Optional): Desired data type (e.g., `int`, `bool`, `float`).
- `order` (Optional): 'C' for row-major or 'F' for column-major storage.

```python
# Create a 1D array of 5 ones (floats)
np.ones(5)
```

```python
# Create a 2D array (matrix) of 3x2 ones (integers)
np.ones((3, 2), dtype=int)
```
